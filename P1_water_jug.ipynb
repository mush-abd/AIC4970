```python
import queue

class WaterJug:

    def __init__(self, capacity_a, capacity_b):
        self.cap_a = capacity_a
        self.cap_b = capacity_b
        self.state = [0, 0]
        self.visited = set()
        self.queue = queue.Queue()

    def is_goal(self, goal):
        return self.state == goal
    
    def calc_vol(self, a):

        if a > max(self.cap_a, self.cap_b):
            return -1
        
        self.queue.put([0, 0])
        
```

```python
from collections import deque


def water_jug(cap_a, cap_b, d):

    if d > max(cap_a, cap_b):
        return -1 

    q = deque([(0, 0)])
    
    visited = [[False] * (cap_b + 1) for _ in range(cap_a + 1)]  
    visited[0][0] = True

    print(f'Queue: {q}')

    while q:
        jug1, jug2 = q.popleft()

        if jug1 == d or jug2 == d:
            print(f"Reached target: {d} litres")
            print(f"Final state: ({jug1}, {jug2})")

            return

        # 1: Fill jug1
        if not visited[cap_a][jug2]:
            visited[cap_a][jug2] = True
            print(f"Filling jug1: {cap_a} litres")
            print(f"state: ({cap_a}, {jug2})")
            q.append((cap_a, jug2))

        # 2: Fill jug2
        if not visited[jug1][cap_b]:
            visited[jug1][cap_b] = True
            print(f"Filling jug2: {cap_b} litres")
            print(f"state: ({jug1}, {cap_b})")
            q.append((jug1, cap_b))

        # 3: Empty jug1
        if not visited[0][jug2]:
            visited[0][jug2] = True
            print(f"Emptying jug1")
            print(f"state: (0, {jug2})")
            q.append((0, jug2))

        # 4: Empty jug2
        if not visited[jug1][0]:
            visited[jug1][0] = True
            print(f"Emptying jug2")
            print(f"state: ({jug1}, 0)")
            q.append((jug1, 0))

        # 5: Pour jug1 into jug2
        pour1to2 = min(jug1, cap_b - jug2)
        if not visited[jug1 - pour1to2][jug2 + pour1to2]:
            visited[jug1 - pour1to2][jug2 + pour1to2] = True
            print(f"Pouring jug1 into jug2: {pour1to2} litres")
            print(f"state: ({jug1 - pour1to2}, {jug2 + pour1to2})")
            q.append((jug1 - pour1to2, jug2 + pour1to2))

        # 6: Pour jug2 into jug1
        pour2to1 = min(jug2, cap_a - jug1)
        if not visited[jug1 + pour2to1][jug2 - pour2to1]:
            visited[jug1 + pour2to1][jug2 - pour2to1] = True
            print(f"Pouring jug2 into jug1: {pour2to1} litres")
            print(f"state: ({jug1 + pour2to1}, {jug2 - pour2to1})")
            q.append((jug1 + pour2to1, jug2 - pour2to1))

        print(f'Queue: {q}')

    return -1  

if __name__ == "__main__":
  
    # jug1 = 4 litre, jug2 = 3 litre 
    # cap_a, cap_b, d = 4, 3, 2
    cap_a = int(input("Capacity of jug1"))
    cap_b = int( input("Capacity of jug2: "))
    d = int(input("Amount of water to calculate:"))

    print(f"Jug1 capacity: {cap_a} litres, Jug2 capacity: {cap_b} litres, Target: {d} litres")

    (water_jug(cap_a, cap_b, d))
```
